## Difference-in-differences with variation in treatment timing

Nosso ponto de partida é a especificação padrão do TWFE, a qual é dada por:

$$ Y_{i,t} = \mu_{i} + \lambda_{t} + \beta D_{i,t} + \gamma X_{i,t} + \epsilon_{i,t} $$

Em todos os exemplos anteriores estávamos representando o chamado DD-Canônico, isto é $D_{it} $ indicava os tratados a partir de um único período específico. Entretanto, essa especificação padrão pode representar, também, um modelo ainda mais flexível, com o tratamento iniciando em momentos diferentes (variation in treatment timing) para diferentes grupos tratados. 

Com o início do tratamento variando entre os grupos de indivíduos tratados (heterogeneidade temporal do tratamento), surgiu uma pergunta: **O que, de fato, o TWFE estima quando existem efeitos de tratamento heterogêneos no tempo?**


### Overview sobre as desconfianças quanto ao estimador TWFE com variação no momento do tratamento

A robustez do estimador TWFE tem sido questionada em estudos que envolvem variação no momento do tratamento. Diversos trabalhos atacaram essa questão. Em dois contextos, (i) no ambiente de estudos de eventos e (ii) na mensuração do $\beta$ ATT com efeito heterogêneo no tempo.

**Event Studies:**
* Borusyak and Jaravel (2017) - Revisiting Event Study Designs, with an Application to the Estimation of the Marginal Propensity to Consume - (Working Paper)
  * [...] a principal mensagem para os profissionais é que, devido a questões de identificação e à ponderação negativa nos desenhos de estudos de eventos, os resultados de especificações comuns provavelmente parecerão não robustos. Estes problemas podem ser atenuados de uma forma baseada em princípios, através da utilização de estimadores e testes paramétricos e semiparamétricos.

* Sun and Abraham (2021), “Estimating Dynamic Treatment Effects in Event Studies with Heterogeneous Treatment Effects,” Journal of Econometric;
  * "Para estimar os efeitos dinâmicos de um tratamento "absorbing", os pesquisadores frequentemente utilizam regressões TWFE que incluem lags e leads do tratamento. Mostramos que, em configurações com variação no momento do tratamento entre unidades, o coeficiente em um determinado lag ou lead pode ser contaminado por efeitos de outros períodos, e pretendências aparentes podem surgir apenas da heterogeneidade dos efeitos do tratamento. Os autores propõe um estimador alternativo que está livre de contaminação e ilustramos as deficiências relativas das regressões de efeitos fixos de duas vias com avanços e atrasos por meio de uma aplicação empírica.

**$\beta$ ATT com efeito heterogêneo no tempo:**

* de Chaisemartin and d’Haultfoeuille (2020), “Two-Way Fixed Effects Estimators with Heterogeneous Treatment Effects,” American Economic Review;
  * "[...] Mostramos que **elas estimam somas ponderadas** dos efeitos médios do tratamento (ATE) em cada grupo e período, **com pesos que podem ser negativos**. Devido aos pesos negativos, o coeficiente da regressão linear pode, por exemplo, ser negativo enquanto todos os ATEs são positivos. Propomos outro estimador que resolve esse problema. Nas duas aplicações que revisitamos, ele é significativamente diferente do estimador de regressão linear."

* Goodman-Bacon (2021), “Difference-in-Differences with Variation in Treatment Timing,” Journal of Econometrics
  * "[...] **o TWFE é igual a uma média ponderada de todos os possíveis estimadores DD de dois grupos/dois períodos nos dados**. Uma interpretação causal dos estimadores de DD de efeitos fixos de duas vias requer tanto uma suposição de tendências paralelas quanto efeitos de tratamento constantes ao longo do tempo. Eu mostro como decompor a diferença entre duas especificações e forneço uma nova análise de modelos que incluem controles variáveis no tempo."

As figuras (da A à D) mostram na linha vertical em vermelho o momento do tratamento. Em triângulos e círculos pretos temos os grupos tratados. Em cinza temos o grupo dos nunca tratados.

É possivel notar que surgem as seguintes formas de construir grupos de comparação:

1. O uso de grupos **nunca tratados (Untreated group)** e que **ainda não não foram tratados (Late group)**, como o grupo de comparação (A, B e C na figura), e 
2. O uso de grupos que foram **tratados em períodos anteriores** (D na figura) como o grupo de comparação. 

* A primeira possibilidade é exatamente o tipo de **comparação que deve ser usado** nas aplicações de Diff-In-Diff (A, B e C).
* O segundo tipo de comparação (às vezes chamado de **“comparações ruins”** ou **“comparações proibidas”**) geralmente não é desejável nessas aplicações e pode levar a estimativas ruins de parâmetros de efeito causal, particularmente na presença de dinâmica de efeito de tratamento.

Ou seja, **o DD tradicional baseado no TWFE (Two-Way Fixed Effects) quando há tratamento em diferentes momentos do tempo é uma caixa preta**, onde ele pode realizar a regressão e ponderar a estimativa de forma a obter um efeito equivocadado, podendo gerar até mesmo pesos negativos para os diferentes grupos de unidades consideradas na análise. Assim, **esse tipo de regresseão não é robusta à existência de heterogeneidade temporal do tratamento**.

**Goldman-Bacon (2021) decompõe o estimador TWFE em diversos DD canônicos e mensura o peso que o TWFE atribui a cada comparação para que chegue ao resultado final.**

No artigo, Goldman-Bacon (2021) replica a análise de:

### Verificando o Viés e a inconsistência do TWFE com tratamentos em diferentes momentos no tempo em Python

```python
# Pacotes
import numpy as np
import pandas as pd
import plotnine as p
import statsmodels.api as sm
import linearmodels as lm
import statsmodels.formula.api as smf 
from linearmodels.panel import PanelOLS
from statsmodels.iolib.summary2 import summary_col
```

```python
# Carregar a base de dados
data = pd.read_stata("https://github.com/Daniel-Uhr/data/raw/main/bacon_example.dta")
```

```python
# Filtragem dos dados
# Vamos criar identificadores estaduais
data['id'] = data['stfips'].rank(method='dense').astype(int)

# Criar uma nova variável de tempo "time"
data['time'] = data['year'] - 1963

# Outcome (Suicide Mortality)
data['Y'] = data['asmrs']
# Treatment
data['D'] = data['post']
# Covariáveis - pcinc asmrh cases
data['X1'] = data['pcinc']
data['X2'] = data['asmrh']
data['X3'] = data['cases']

data['_nfd'].describe()

data['media_D'] = data.groupby('id')['D'].transform('mean')
data['always_treated'] = np.where(data['media_D'] == 1, 1, 0)
data['never_treated'] = np.where(data['media_D'] == 0, 1, 0)
```

```python
# Regressão controlando os efeitos fixos id e ano, com erro padrão robusto à heterocedasticidade.
reg1 = smf.wls('Y ~ D + C(id) + C(year)', data=data).fit(cov_type='HC1')
#print(reg1.summary())

reg1.summary2().tables[1].loc[['Intercept', 'D'], :].round(4)
```

```python
# Restringir o dataframe para _nfd!=.
data_restricted = data[data['_nfd'].notnull()]
# Ajustar a regressão WLS
reg2 = smf.wls('Y ~ D + C(id) + C(year)', data=data_restricted).fit(cov_type='HC1')
# Imprimir o resumo
#print(reg2.summary())
reg2.summary2().tables[1].loc[['Intercept', 'D'], :].round(4)
```

```python
# Regressão controlando os efeitos fixos id e ano, com erro padrão robusto à heterocedasticidade, ponderada pela variável data['weight'].
reg3 = smf.wls('Y ~ D + C(id) + C(year)', data=data, weights=data['weight']).fit(cov_type='HC1')
#print(reg3.summary())
reg3.summary2().tables[1].loc[['Intercept', 'D'], :].round(4)
```

```python
# Regressão controlando os efeitos fixos id e ano, com erro padrão robusto à heterocedasticidade, e covariáveis.
reg4 = smf.wls('Y ~ D + X1 + X2 + X3 + C(id) + C(year)', data=data).fit(cov_type='HC1')
#print(reg4.summary())
reg4.summary2().tables[1].loc[['Intercept', 'D'], :].round(4)
```

```python
# Regressão controlando os efeitos fixos id e ano, e o produto entre efeitos fixos e tempo, com erro padrão robusto à heterocedasticidade.
reg5 = smf.wls('Y ~ D + C(id)*time + C(id) + C(year)', data=data).fit(cov_type='HC1')
#print(reg5.summary())
reg5.summary2().tables[1].loc[['Intercept', 'D'], :].round(4)
```

* Não há como realizar as estimações (4), (7) e (8) da tabela 2 porque não temos dados disponíveis nessa base de dados utilizada.

Nos chama a atenção os diferentes valores encontrados, mostrando uma clara **falta de robustez do estimador TWFE quando o tratamento ocorre em diferentes momentos do tempo** (adoção do tratamento escalonada - **Stagged Adoption**). Assim, Goldman-Bacon argumenta que isso se deve aos diferentes grupos de comparação que o TWFE realiza, atribuindo pesos diferentes para cada um deles. Logo, é necessário que a análise seja feita de forma mais cuidadosa, entendendo os grupos de comparação envolvidos.

Para finalizar, podemos verificar como ficaria uma regressão para uma amostra restrita sem os estados "Always Treated", ou seja, sem os estados que sempre tiveram a intervenção implementada durante o período da base.


```python
# Restringir o dataframe para always_treated!=1
data_restricted2 = data.loc[data['always_treated'] != 1].copy()
reg_rest2 = smf.wls('Y ~ D + C(id) + C(year)', data=data_restricted2).fit(cov_type='HC1')
# Imprimir o resumo
#print(reg_rest2.summary())
reg_rest2.summary2().tables[1].loc[['Intercept', 'D'], :].round(4)
```

vamos comparar os resultados obtidos aqui em uma unica tabela.

```python
import pandas as pd

# Criando uma função para arredondar os valores
def round_results(results):
    return [round(value, 4) if isinstance(value, (int, float)) else value for value in results]

# Criando um DataFrame vazio para armazenar os resultados
results_table = pd.DataFrame(index=['Effect', 'S.E.', 'T-test', 'P-value', 'N.', 'Adj. R-squared'])

# Definindo os modelos e os resultados de interesse
models = [reg1, reg2, reg3, reg4, reg5, reg_rest2]
model_names = ['Model 1', 'Model 2', 'Model 3', 'Model 4', 'Model 5', 'Model 6']

# Loop para iterar sobre os modelos
for model, model_name in zip(models, model_names):
    # Extrair os resultados de interesse
    coef = model.params['D']
    std_err = model.bse['D']
    t_value = model.tvalues['D']
    p_value = model.pvalues['D']
    sample_size = len(model.fittedvalues)
    adj_r_squared = model.rsquared_adj
    
    # Arredondar os resultados
    model_results = round_results([coef, std_err, t_value, p_value, sample_size, adj_r_squared])
    
    # Preencher a tabela com os resultados
    results_table[model_name] = model_results

# Exibir a tabela
print(results_table.to_markdown())
```

### Dadas as inconsistências, novos estimadores foram propostos na literatura para lidar com a heterogeneidade temporal do tratamento.

As pesquisas recentes procuram relaxar uma ou duas suposições do DD:
* Permitir múltiplos períodos e variação no timming do tratamento 
  * novos estimadores isolam comparações “limpas” entre grupos ainda não tratados e, em seguida, agrega-os usando pesos especificados pelo usuário para estimar.
* Considerar a violação potencial das tendencias paralelas
  * São tendências paralelas quando condicionais às covariáveis observadas.
* Partir da suposição de observar uma amostra de muitos clusters independentes de uma população
  * Os métodos padrão de cluster-robust pressupõem que há um grande número de clusters tratados e não tratados e, portanto, podem ter um desempenho ruim no caso de um pequeno cluster tratado. Uma variedade de alternativas com melhores propriedades foram propostas para este caso, incluindo procedimentos de permutação e bootstrap.

* **Estimadores robustos à Heterogenidade temporal do tratamento**

### Callaway & Sant´Anna (2021)

A partir dos pontos levantados por Goldman-Bacon (2021), os autores Brantly Callaway & Pedro Sant´Anna se propõem a construir um estimador de diferença em diferenças robusto para o caso de tratamentos em diferentes momentos do tempo.

**Novidades do artigo**

* Permitem covariáveis de forma flexível
* Procedimento de estimação baseado em Doubly Robust Methods (combinação de Outcome Regression DD, e IPW DD)
* Discutem diferentes formas de agregação dos dados para sumarizar o efeito do tratamento;
* Cobrem tanto painel quanto cross-sections repetidos
* Procuram fazer hipóteses mínimas de tendências paralelas para identificar o ATT no grupo afetado em determinado ano (ATT(g,t)).

**Contexto geral do artigo**

Em termos procedimentais, os autores resolvem o problema dos *tratamentos escalonados* no tempo (**Stagged Adoption**) ao desagregar os dados em diversos DD canônicos 2x2. Eles buscam identificar o momento específico em que alguns indivíduos são tratados, e separam a análise para blocos específicos ou “bloco de construção” ("building blocks"). 

Por exemplo, indivíduos que foram tratados no ano 7, são agrupados (Grupo: g=7), e é calculado o efeito do DD entre esse grupo (Coorte/Cohort) específico e o grupo de controle (Controle: "nunca tratados", ou "ainda não tratados"). Adicionalmente, os autores consideram o "Escore de Propensão Generalizado" como um peso para cada bloco de construção. Assim, combinam as estratégias de *Outcome Regression* de Heckman et al.(1997, 1998) com a de escore de propensão proposta por Abadie (2005) para obter uma configuração de DD com múltiplos períodos, este procedimento de estimação é denominado *duplamente robusto*.

Ao separ a análise em blocos de construção 2x2, eles obtém outras métricas do efeito do tratamento. Pode-se obter o ATT geral, e seus componentes:

* Efeito médio overall (simple) [$ATT_{Simple}$]
* Efeito médio por Grupos (Coortes/Cohort) [$ATT_G$]
* Efeito Calendário (efeito agregado em cada ano t) [$ATT_C$]
* Efeito Dinâmico (Dynamic / Event Study Design) [$ATT_D$]


**Hipóteses (Assumptions)**

* Tendências Paralelas Condicionais (Conditional Parallel Trends - para os nunca tratados [Never treated, $C=1$] ou não ainda tratados)

$$ E[Y_{t}(0) - Y_{t-1}(0)|X, G_{g}=1] = E[Y_{t}(0) - Y_{t-1}(0)|X, C=1] $$

* Não antecipação do tratamento (No anticipation of treatment)
* Tratamento irreversível
* Suporte Comum (Common Support) - sobreposição das covariáveis para o propensity score
* Amostra é i.i.d.

**Parâmetro de Interesse**

O "bloco de construção" é dado por:

$$ ATT(g,t) = E[Y_{t}(g) - Y_{t}(0)|G_{g}=1] $$

Para $t$ maior ou igual à g. Depois é possível agregar esses blocos entre as Cohort, no tempo, ou na combinação dos dois. É uma generalização natural do parâmetro ATT 2x2.

O primeiro termo é observado ($Y_{t}(g)$), o segundo é o contrafactual.


**Procedimento geral**

* Aplicação do DD canônico 2x2 para os diferentes Cohorts (Grupos) em seus anos específicos.
* Ou seja, precisamos apenas de um ano antes e um depois da ocorrência do tratamento para cada grupo.
* OBS: Sem covariáveis, o ATT(g,t) é semelhante ao estimador de de Chaisemartin & D’Haultfoeuille (2022), e Sun & Abraham (2021).
* Adicionando as Covariáveis, o método muda completamente.


#### Callaway & Sant´Anna (2021) Doubly Robust Difference-in-Differences 

**Modelo Geral**

Com uma estrutura de dados em painel balanceado, o ATT para o grupo ($g$) no período ($t$), e grupo de controle com os *never-treated* ($C=1$), é estimado de forma semi-paramétrica como:

$$ ATT(g,t)=E \left [ \left ( \frac{G_{g}}{E[G_{g}]}  - \frac{\frac{p_{g,t}(X)C}{1-p_{g,t}(X)}}{E[\frac{p_{g,t}(X)C}{1-p_{g,t}(X)}]} \right ) (Y_{t}-Y_{g-1} - \mu_{g,t}^{nev}(X)) \right ]$$

onde 

$$\mu_{g,t}^{nev}(X) = E[Y_{t}-Y_{g-1}|X, C=1]$$

Esse estimador extende Heckman, Ichimura and Todd (1997), Abadie (2005), Sant´Anna and Zhao (2020).
* **Primeira parte**: Temos 2 grupos, a Cohort $G_{g}$ e/ou a Cohort never-treated $C$. As duas são variáveis *binárias*. Se as duas são zero, descartamos todos os dados. Os denominadores $E[G_{g}]$ e $E[\frac{p_{g,t}(X)C}{1-p_{g,t}(X)}]$ estão lá para garantir que esses pesos se integrem até 1 (são "pesos estabilizados"). Esses estabilizadores ajudam a não permitir que os valores subam muito caso o escore de propensão seja muito próximo de 1. O uso do propensity score busca atribuir mais peso àqueles grupos de controle mais semelhantes às cohorts de g. O propensity score é a probabiliade de ser tratado no período t, dado que ainda não foi tratado.
* **Segunda parte**: Estamos fazendo o DD, considerando a evolução do resultado do grupo de controle condicional nas covariáveis (Outcome Regression).
* Como estamos combinando o IPW com o Outcome Regression, o estimador é duplamente robusto. Ou seja, se uma das duas partes estiver correta, o estimador será consistente, e recuperará o ATT. "Temos dois tiros para acertar o alvo".

**Sumarizando os ATT(g,t)´s**

Os autores propõem tomar a média ponderada dos ATT(g,t)´s.

* Análise ponderada dos ATT(gt)´s, aqui começa a brincadeira, e há certa liberdade para escolher os pesos (w). Ou seja, esses pesos podem assumir diferentes formas, dependendo do objetivo da análise.

$$\sum_{g=2}^{\tau} \sum_{t=2}^{\tau} w_{g,t} ATT(g,t)$$


* **ATT Simple** (simple weighted average / overall average): Pesos iguais para todos (w=1) para todos. Pode ser ruim porque, umas cohorts são maiores que outras.

$$\theta_{M} = \frac{2}{\tau(\tau-1)}\sum_{g=2}^{\tau} \sum_{t=2}^{\tau} 1[g \leqslant t] ATT(g,t)$$

  * Entendendo o *ATT Simple*:
    * $\tau$ é o número total de períodos no estudo (horizonte temporal da análise).
    * $\frac{2}{\tau(\tau-1)}$ é um fator de normalização que assegura que a soma dos pesos atribuídos seja igual a 1, independentemente do número de grupos e períodos.
    * A soma dupla percorre todos os grupos/coortes $g$ que começam a ser tratados a partir do segundo período ($g=2,3,...,\tau$) e todos os períodos $t$ do calendário.
    * $1[g \leqslant t]$ é uma função indicadora que é 1 se g é menor ou igual a t, e 0 caso contrário.
    * OBS: Caso o tratamento ocorresse em $g=1$ não seria possível calcular o ATT, pois não teríamos um período anterior para comparação.

**ATT(g,t)´s que exploram as heterogeneidades**

  * **ATT by Group**: Efeito Médio de participação no tratamento que as unidades no grupo g experimentam.
     * Enfatiza a heterogeneidade entre os grupos.

$$ \theta_{S}(g) = \frac{1}{\tau - g + 1} \sum_{t=2}^{\tau} 1[g \leqslant t ]ATT(g,t)$$

  * Entendendo o *ATT by Group*:
    * $\theta_{S}(g)$ é o efeito médio do tratamento para **o grupo específico** $g$ (ou coorte $g$). Mede como as unidades do grupo $g$ foram impactadas pelo tratamento ao longo dos períodos em que o tratamento foi implementado.
    * $\frac{1}{\tau - g + 1}$: É um fator de normalização que divide a soma pelo número de períodos $t$ válidos para o grupo $g$.
    * $\sum_{t=2}^{\tau}$ é a soma dos períodos em que o tratamento foi implementado. Soma todos os períodos $t$, começando em $t=2$ até o último período $\tau$.
    * $1[g \leqslant t ]$ 
    * Função indicadora que vale 1 se o período $t$ ocorre no mesmo ano ou após o início do tratamento para o grupo $g$, ou 0 caso contrário. Isso garante que apenas os períodos em que o grupo $g$ já foi tratado (ou está em sua janela de pós-tratamento) sejam considerados no somatório.


  * **ATT by Calendar Period**: Efeito médio de participação no tratamento no período t para grupo g que participaram no tratamento a partir do período t. Lembra um Event Study individual para cada grupo g. E posteriormente, é feita uma média ponderada desses efeitos.
    
$$ \theta_{C}(t) = \sum_{t=2}^{\tau} 1[g \leqslant t ]ATT(g,t) P(G=g|C \neq 1)$$


  * Entendendo o *ATT by Calendar Period*:
    * $\tau$ é o número total de períodos no estudo (horizonte temporal da análise).
    * $\sum_{t=2}^{\tau}$ é a soma dos períodos em que o tratamento foi implementado. Soma todos os períodos $t$, começando em $t=2$ até o último período $\tau$.
    * $1[g \leqslant t ]$ é uma função indicadora que é 1 se g é menor ou igual a t, e 0 caso contrário. Essa função garante que apenas os grupos tratados no período $g$ ou antes de $t$ contribuam para o cálculo.
    * $P(G=g|C \neq 1)$ representa a proporção de unidades pertencentes ao grupo $g$ entre as que não são "never-treated". É um peso que ajusta a contribuição relativa de cada grupo $g$ para o cálculo do $\theta_{C}(t)$, garantindo que grupos maiores tenham maior influência no resultado agregado.
    * O $\theta_{C}(t)$ mede o efeito médio do tratamento observado no período $t$, combinando informações de todos os grupos tratados até aquele momento. Ele captura a heterogeneidade dos efeitos entre os grupos no mesmo período calendário.


  * **ATT Dynamic**: Pode-se reparametrizar a análise para avaliar o efeito médio em relação a um momento específico (e) a partir do início do tratamento (Event Study).
    
$$\theta_{D}(e) = \sum_{g=2}^{\tau} 1[g + e \leqslant \tau] ATT(g, g + e) P(G=g| G + e \leqslant \tau, C \neq 1)$$

  * Entendendo o *ATT Dynamic*:
    * $e$ é o número de períodos decorridos desde o início do tratamento $g$. Também chamado de *tempo relativo* ou *momento do evento* (em um event study).
    * $\tau$ é o número total de períodos no estudo (horizonte temporal da análise).
    * $g$ é o grupo tratado - Ano de início do tratamento para um grupo específico.
    * $1[g + e \leqslant \tau]$ é uma função indicadora que é 1 se o período relativo ao início do tratamento $(g+e)$ ainda está dentro do horizonte temporal do estudo ($g+e \leqslant \tau$), ou 0 caso contrário. Isso garante que a soma inclua apenas os períodos em que os efeitos ainda podem ser observados.
    * ATT(g, g + e) é o efeito do tratamento no período relativo $e$ (isto é, no período calendário $g + e$) para o grupo que começou a ser tratado em $g$. Mede como o tratamento afeta o grupo $g$ e períodos após o início do tratamento.
    * $P(G=g| G + e \leqslant \tau, C \neq 1)$ representa a proporção de unidades pertencentes ao grupo $g$ entre as que ainda podem contribuir para a análise no período $g+e$($g+e \leq \tau$) .


## Aplicação do estimador de Callaway & Sant´Anna (2021) em Python

Existe um pacote no python para realizar as estimações propostas por Callaway & Sant´Anna (2021). O pacote é o **differences** está baseado nos artigos:

* Sant’Anna, Pedro HC, and Jun Zhao. “Doubly robust difference-in-differences estimators.” Journal of Econometrics 219, no. 1 (2020): 101-122.
* Callaway, Brantly, and Pedro HC Sant’Anna. “Difference-in-differences with multiple time periods.” Journal of Econometrics 225, no. 2 (2021): 200-230.
* Callaway, Brantly, Andrew Goodman-Bacon, and Pedro HC Sant’Anna. “Difference-in-differences with a continuous treatment.” arXiv preprint arXiv:2107.02637 (2021). [multi-valued treatment case]

Ver: https://github.com/bernardodionisi/differences 

OBS: O pacote permite o cálculo de tripla-differença, tratamento para multiplos-valores.


Vamos utilizar a base de dados do artigo de Stevenson, Betsey, Wolfers, Justin, 2006. Bargaining in the shadow of the law: Divorce laws and family distress. Q. J. Econ. 121 (1), 267–288.

```python
# Pacotes
import pandas as pd
from differences import ATTgt
import numpy as np
```

Carregando a base de dados.

```python
data = pd.read_stata("https://github.com/Daniel-Uhr/data/raw/main/bacon_example.dta")
```

Filtrando e construindo as variáveis necessárias para a estimação.

```python
# Filtragem dos dados
# Vamos criar identificadores estaduais
data['id'] = data['stfips'].astype('category').cat.codes + 1

# Outcome (Suicide Mortality)
data['Y'] = data['asmrs']
# Treatment
data['D'] = data['post']
# Covariáveis - pcinc asmrh cases
data['X1'] = data['pcinc']
data['X2'] = data['asmrh']
data['X3'] = data['cases']

# Vamos criar a variável de grupo G
data['G']=data['_nfd']
```

```python
# O pacote precisa entender a estrutura de painel dos dados. Precisamos ajustar a variável year para int (inteiro)
# Criando clones
data['year1'] = data['year']
data['id1'] = data['id']
data['year1'] = data['year'].astype(int)

# Definir os indices (estrutura de painel)
data.set_index(['id1', 'year1'], inplace=True)
```

A variável mais importante para utilização desse método é a criação da variável de identificação do ano de inicio do tratamento para cada grupo ($g$), que vamos chamar de "G". Além disso, esse método requer que definamos os índices de identificador de tempo e indivíduo (declarar a estrutura de painel de dados para o sistema - definição dos índices).

```python
tabulacao_data = data['G'].value_counts().reset_index()
tabulacao_data.columns = ['G', 'Contagem']
print(tabulacao_data)
```

Os dados mostram que temos os seguintes grupos:

* 1969 
* 1970

Ou seja, se pensarmos em termos individuais, é como se tivéssemos 12 análises de diferenças em diferenças individuais possíveis. Dessa forma, Callaway e Sant´Anna (2021) seguem ampliam a abordagem de Sant´Anna e Zhao (2020), ao pensar o DD em termos de "blocos de construção" considerando mais períodos de tempo.

Agora, já podemos realizar as estimações. Primeiro vamos indicar o dataframe, e a variável de identificação do grupo de tratamento. Em seguida, vamos realizar a estimação do modelo.

```python
att_gt = ATTgt(data=data, cohort_name="G")
```

Agora vamos estimar o modelo. Repare que é possivel definir a fórmula. E também indicar se queremos realizar "dr" o Doubly Robust (logistic propensity score model for the probability of being treated); ou "dr-ipt" (with propensity score estimated using the inverse probability tilting), "reg" (for outcome regression DiD estimator), "std_ipw-mle" (for standardized inverse probability weighted DiD estimator, with logistic propensity score model for the probability of being treated).

```python
att_gt.fit("Y ~ X1 + X2", est_method="dr")
```

**Agregação Simples (Overall ATT averages)**

Uma ideia que provavelmente virá imediatamente à mente é apenas retornar uma média ponderada de todos os efeitos médios do tratamento no tempo do grupo, **com pesos proporcionais ao tamanho do grupo**. Isso está disponível chamando a aggt-função "simple".

```python
att_gt.aggregate("simple")
```

Esse tipo de agregação evita imediatamente o problema dos pesos negativos que as regressões de efeitos fixos bidirecionais podem enfrentar, mas muitas vezes pensamos que existem alternativas melhores . Em particular, esta simples agregação tende a sobrecarregar o efeito dos grupos tratados precocemente, simplesmente porque observamos mais deles durante os períodos pós-tratamento. Acreditamos que provavelmente haverá alternativas melhores na maioria das aplicações.

**Efeitos Específicos do Grupo (Cohort)**

Outra ideia é observar os efeitos médios específicos de cada grupo. Também é simples agregar os efeitos médios do tratamento no tempo do grupo em efeitos médios do tratamento específicos do grupo usando o seguinte código:

```python
att_gt.aggregate("cohort")
```

```python
att_gt.aggregate("cohort", overall=True)
```

**Efeito Calendário (Calendar Time Effects)**

para calcular o efeito médio do tratamento em cada "momento de tempo"

```python
att_gt.aggregate("time")
```

```python
att_gt.aggregate("time", overall=True)
```

**Efeitos Dinâmicos ou Estudos de Eventos**

Uma das abordagens alternativas mais comuns é agregar os efeitos do tempo do grupo em um gráfico de estudo de eventos. Os efeitos médios do tratamento no tempo de grupo podem ser imediatamente calculados em termos de efeitos médios do tratamento em diferentes durações de exposição ao tratamento usando o seguinte código:

```python
att_gt.aggregate("event")
```

```python
att_gt.aggregate("event", overall=True)
```

Vamos salvar os resultados do Event Study em um dataframe. Para criar um gráfico de Event Study.

```python
es_attgt_res = att_gt.aggregate("event")
es_attgt_res = es_attgt_res.reset_index()  # Transforma o índice em uma coluna
es_attgt_res['relative_period'] = pd.to_numeric(es_attgt_res['relative_period'])  # Converte para numérico
```

```python
# Remover as duas primeiras linhas do DataFrame
es_attgt_res_clean = es_attgt_res.iloc[2:].reset_index(drop=True)
# Renomear as colunas para os nomes corretos
es_attgt_res_clean.columns = ['relative_period', 'ATT', 'std_error', 'lower', 'upper', 'zero_not_in_cband']
es_attgt_res_clean
```

Plotando o gráfico de Event Study.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar o tamanho do gráfico
plt.figure(figsize=(10, 6))

# Plotar os coeficientes com barras de erro
plt.errorbar(
    es_attgt_res_clean['relative_period'],  # Valores do eixo x
    es_attgt_res_clean['ATT'],   # Valores do eixo y (coeficientes)
    yerr=[
        es_attgt_res_clean['ATT'] - es_attgt_res_clean['lower'], 
        es_attgt_res_clean['upper'] - es_attgt_res_clean['ATT']
    ],  # Barras de erro
    fmt='o', color='black', markersize=4, ecolor='gray', elinewidth=0.8, capsize=2, label='Estimated Coefficients'
)

# Adicionar linhas horizontais para valores de referência
plt.axhline(-5.940387, color='blue', linestyle='-', linewidth=1, label='Overall (-5.9404)')
plt.axhline(0, color='black', linestyle='--', linewidth=0.8, label='Baseline (0)')

# Adicionar uma linha vertical para marcar o evento
plt.axvline(0, color='black', linestyle='--', linewidth=0.8)

# Adicionar título e rótulos com fontes menores
plt.title('Years Before and After Treatment', fontsize=12)
plt.xlabel('Years Before and After Treatment', fontsize=10)
plt.ylabel('Coefficient Estimate (log(Homicide Rate))', fontsize=10)

# Ajustar legenda
plt.legend(frameon=False, fontsize=10)

# Remover linhas superiores e à direita do gráfico
sns.despine()

# Exibir o gráfico
plt.show()
```

### Sun & Abraham (2021)

Sun e Abraham (2021) propõem um estimador alternativo para estudos de eventos que aborda os problemas de **heterogeneidade nos efeitos do tratamento** em cenários com tratamentos escalonados (*staggered adoption*). A proposta visa solucionar limitações das regressões tradicionais de efeitos fixos (*Two-Way Fixed Effects - TWFE*), que podem gerar estimativas enviesadas em situações com tratamentos heterogêneos.

**Problemas com Regressões TWFE**
- O uso de regressões TWFE em cenários com tratamentos escalonados pode gerar:
  - **Contaminação nas estimativas**: Coeficientes podem refletir efeitos de períodos não relacionados.
  - **Pré-tendências falsas**: Falsos positivos indicando mudanças antes do tratamento.
- Esses problemas comprometem a interpretação dos resultados.

**Necessidade de Alternativas**
- É necessário um método que seja:
  - **Robusto à heterogeneidade dos efeitos do tratamento.**
  - **Capaz de fornecer estimativas claras e interpretáveis.**


### **Estimador Alternativo: Interaction-Weighted Estimator (IW)**

O estimador IW baseia-se na criação de médias ponderadas dos efeitos específicos de cada coorte.

#### 1. **CATT: Cohort Average Treatment Effect on the Treated**
- A métrica base do estimador é o **Cohort Average Treatment Effect on the Treated (CATT)**:
$$
  CATT_{e, \ell} = E[Y_{i,e+\ell} - Y_{i,1} | E_i = e]
$$
  - $ e $: Coorte de tratamento.
  - $ \ell $: Tempo relativo ao tratamento.
  - $ Y_{i,e+\ell} $: Resultado no tempo $ e+\ell $.

O CATT(e,l) é uma reformulação do $ ATT(g,t) $ de Callaway e Sant'Anna, mas focado na dinâmica relativa ao tratamento. Ambos medem efeitos médios de tratamento, mas o $ CATT(e,l) $ facilita a análise dinâmica (ex.: pré-tratamento e pós-tratamento).

A equação representa o Cohort Average Treatment Effect on the Treated (CATT), que é o efeito médio do tratamento em um tempo relativo ($l$) para uma coorte específica ($e$).

* Exemplo: se $e=2$, isso significa que os indivíduos dessa coorte começaram o tratamento no ano 2. O tempo relativo ($l$) é o número de períodos antes ($l<0$), ou após ($l>0$) do início do tratamento para a coorte $e$. Então, se $e=2$ e $l=1$, estamos analisando o impacto um período após o início do tratamento, ou seja, no Ano 3 ($e+l=2+1=3$).
 

#### 2. **Estimador IW**
- O estimador IW calcula uma média ponderada dos $ CATT_{e, \ell} $, usando como pesos as proporções das coortes:
  $$
  \nu_g = \frac{1}{|g|} \sum_{\ell \in g} \sum_{e} CATT_{e, \ell} \cdot \text{Pr}(E_i = e | E_i \in [-\ell, T-\ell])
  $$
  Onde:
  - $ g $: Período relativo ao tratamento.
  - $ \text{Pr}(E_i = e) $: Proporção de indivíduos na coorte $ e $.


O estimador IW calcula o efeito médio ponderado do tratamento para um intervalo específico de tempos relativos ($g$), ajustando as contribuições de cada coorte ($e$) pela sua representatividade ($Pr(E_i=e)$).

**Exemplo prático 1**
* Cenário: Temos três coortes ($e=2,4,6$) e queremos estimar o efeito médio do tratamento no tempo relativo $g=-1$ (um período antes do tratamento). 
* Dados Simulados:
  * $CATT_{2,-1} = -5$
  * $CATT_{4,-1} = -3$
  * $CATT_{6,-1} = -2$
* Proporções das Coortes:
  * $Pr(E_i=2) = 0.4$
  * $Pr(E_i=4) = 0.3$
  * $Pr(E_i=6) = 0.3$
* Cálculo do Estimador IW:
  * $\nu_{-1} = \frac{1}{1} \left( (0.4 \times (-5)) + (0.3 \times (-3)) + (0.3 \times (-2)) \right) = -3.5$

Repare que $|g|$ é o número de tempos relativos incluídos dentro do conjunto g. No nosso caso, $|g|=1$ porque estamos interessados apenas no tempo relativo $g=-1$.


**Exemplo prático 2**

Imagine que estamos interessados em estimar o efeito médio do tratamento nos períodos relativos $−1$ (um período antes do tratamento) e $+1$ (um período após o tratamento). Assim, o conjunto de tempos relativos seria: 

$$ g = \{-1, 1\} $$

Neste caso:
* $|g| = 2$, porque g contém dois tempos relativos.


**Dados Simulados**

Valores do $CATT_{e,\ell}$ (efeitos médios para cada coorte e tempo relativo):

| Coorte ($e$) | $CATT_{e,-1}$ | $CATT_{e,+1}$ |
|--------------|---------------|---------------|
| $e = 2$      | $-5$          | $10$          |
| $e = 4$      | $-3$          | $8$           |
| $e = 6$      | $-2$          | $6$           |

Proporções das Coortes ($\text{Pr}(E_i = e)$):

| Coorte ($e$) | $\text{Pr}(E_i = e)$ |
|--------------|----------------------|
| $e = 2$      | $0.4$               |
| $e = 4$      | $0.3$               |
| $e = 6$      | $0.3$               |


**Estimador IW**

A fórmula geral para o estimador IW é:
$$
\nu_g = \frac{1}{|g|} \sum_{\ell \in g} \sum_{e} CATT_{e, \ell} \cdot \text{Pr}(E_i = e)
$$

Neste exemplo:
- $g = \{-1, +1\}$
- $|g| = 2$

1. Para $\ell = -1$:
$$
\sum_{e} CATT_{e,-1} \cdot \text{Pr}(E_i = e) = (-5 \cdot 0.4) + (-3 \cdot 0.3) + (-2 \cdot 0.3)
$$
$$
= -2.0 - 0.9 - 0.6 = -3.5
$$

2. Para $\ell = +1$:
$$
\sum_{e} CATT_{e,+1} \cdot \text{Pr}(E_i = e) = (10 \cdot 0.4) + (8 \cdot 0.3) + (6 \cdot 0.3)
$$
$$
= 4.0 + 2.4 + 1.8 = 8.2
$$


**Passo 2: Combinar os efeitos ponderados dos períodos relativos**

Agora, calculamos a média dos efeitos ponderados para $\ell = -1$ e $\ell = +1$:
$$
\nu_g = \frac{1}{|g|} \sum_{\ell \in g} \text{(efeito ponderado para } \ell\text{)} = \frac{1}{2} \cdot (-3.5 + 8.2)
$$
$$
\nu_g = \frac{1}{2} \cdot 4.7 = 2.35
$$

O efeito médio ponderado do tratamento para o conjunto de períodos relativos $g = \{-1, +1\}$ é:
$$
\nu_g = 2.35
$$

- O estimador IW combina os efeitos nos períodos relativos $-1$ (pré-tratamento) e $+1$ (pós-tratamento), ajustando para a proporção de indivíduos em cada coorte.
- Isso permite explorar como os efeitos do tratamento se distribuem em **múltiplos períodos relativos**, oferecendo uma visão mais ampla e dinâmica do impacto do tratamento.





#### 3. **Vantagens do Estimador IW**
- **Pesos convexos**: Os pesos são normalizados e somam a 1.
- **Robustez**: Evita contaminação de efeitos não relacionados.
- **Interpretação clara**: Resultados são apresentados como médias ponderadas por coorte e tempo relativo.

### **Comparação com TWFE**

#### 1. **Heterogeneidade nos Efeitos**
- TWFE assume homogeneidade nos efeitos do tratamento:
  - Estimativas podem ser enviesadas se os efeitos do tratamento forem heterogêneos.
- IW permite explorar heterogeneidades específicas para cada coorte e período.

#### 2. **Interpretação**
- O estimador TWFE combina diferentes efeitos em um único coeficiente, dificultando a interpretação.
- IW apresenta resultados desagregados, fornecendo uma visão mais detalhada e intuitiva.

### **Considerações Finais**

1. **Hipóteses Necessárias**
   - **Tendências paralelas**: Suposição de que, na ausência de tratamento, os grupos seguiriam trajetórias similares.
   - **Ausência de antecipação do tratamento**: Os efeitos não podem ocorrer antes da aplicação do tratamento.

2. **Conexão com Outros Métodos**
   - O estimador IW é uma extensão natural de Callaway e Sant'Anna (2021), mas com simplificações que tornam o método mais acessível.

3. **Aplicabilidade**
   - Ideal para cenários em que o tratamento é aplicado em diferentes momentos e há suspeitas de heterogeneidade nos efeitos.


Vantagens: 
*Agregação Flexível:
  * Permite calcular efeitos dinâmicos para intervalos de tempos relativos ($g$). Útil para estudar pré-tendências ($g<0$) ou efeitos ao longo do tempo ($g≥0).	
* Ponderação Apropriada:
  * As contribuições de cada coorte são ajustadas pela sua representatividade ($Pr(E_{i}=e)$).
* Mitigação de Viés:
  * Evita contaminação das estimativas por efeitos de períodos não relacionados.

## Sun & Abraham (2021) - Difference-in-Differences with Multiple Time Periods

As autoras Sun & Abraham (2021) desenvolveram um estimador que aborda os desafios do uso de Diferença em Diferenças (DiD) em cenários com tratamentos em diferentes momentos do tempo (também conhecido como Staggered Adoption). Esse método busca lidar com problemas de heterogeneidade nos efeitos do tratamento e com violações das hipóteses de tendências paralelas em cenários mais complexos.

**Contexto Geral**

Quando os tratamentos são implementados em diferentes momentos do tempo, os métodos tradicionais de DiD, como o estimador interativo de Goodman-Bacon (2021), podem sofrer de contaminação nos efeitos estimados. Isso ocorre porque indivíduos tratados em um período específico podem, inadvertidamente, atuar como grupo de controle em períodos posteriores, comprometendo a identificação causal.

Sun & Abraham propõem uma abordagem para desagregar os efeitos em componentes mais claros e interpretar os resultados de maneira mais intuitiva:

* Cada coorte (grupo tratado no mesmo período) é comparada apenas com um grupo controle válido.
* A estratégia baseia-se em interações entre o momento relativo ao tratamento e as coortes.

Blocos de Construção do Estimador
O estimador se baseia na criação de blocos de construção para analisar os efeitos do tratamento ao longo do tempo. Esses blocos são definidos como:

Coortes (Cohorts): Indivíduos tratados em um mesmo período.
Tempos Relativos ao Tratamento (Relative Time): Representam o número de períodos antes ou depois de o tratamento ter sido aplicado.
Por exemplo:

* Se um indivíduo foi tratado no ano 3, o tempo relativo ao tratamento em relação ao ano 4 será +1, enquanto em relação ao ano 2 será -1.



**Principais Etapas**
* Criação de Indicadores de Tempo Relativo
Os autores criam variáveis para capturar o tempo relativo ao tratamento, permitindo analisar como o efeito do tratamento varia antes e depois da intervenção.

* Controle de Coortes Não Tratadas (Never-Treated)
Para evitar contaminação, o grupo controle é formado por indivíduos que nunca foram tratados ou ainda não receberam o tratamento.

* Interação entre Tempo Relativo e Coorte
Cada coorte é analisada separadamente, permitindo estimar efeitos específicos para diferentes momentos relativos ao tratamento.

* Matriz de Pesos
São usados pesos baseados na proporção de indivíduos em cada coorte, garantindo que a média ponderada dos efeitos reflita o efeito agregado do tratamento.

### Modelo Geral

A regressão estimada inclui:

- **Variáveis dependentes**: Medem o resultado de interesse.
- **Indicadores de tempo relativo**: Capturam o efeito do tratamento em momentos específicos.
- **Pesos**: Ajustam a influência de cada coorte na estimativa final.

O modelo matemático pode ser representado como:

$$
Y_{it} = \alpha + \sum_{k \neq -1} \delta_k \cdot \text{Rel\_Time}_k \cdot \text{Cohort}_g + \varepsilon_{it}
$$

Onde:

- $ \text{Rel\_Time}_k$: Indicadores de tempo relativo.
- $ \text{Cohort}_g $: Indicadores de coorte.
- $ \delta_k $: Efeitos estimados em cada tempo relativo.


### Vantagens do Estimador

#### **Separação Clara dos Efeitos**
O estimador desagrega os efeitos do tratamento para cada tempo relativo, permitindo identificar padrões dinâmicos (e.g., efeitos cumulativos ou decrescentes ao longo do tempo).

#### **Heterogeneidade Explorada**
Os resultados podem ser analisados por coorte, por tempo relativo ou por períodos agregados.

#### **Robustez**
A abordagem reduz os problemas de contaminação que surgem em métodos tradicionais, fornecendo estimativas mais confiáveis.



### Interpretação Intuitiva

Imagine que temos três coortes que receberam o tratamento em anos diferentes (Ano 2, Ano 4 e Ano 6). O estimador de Sun & Abraham funciona como um **"zoom analítico"**:

1. **Foco em cada coorte separadamente**:
   - No Ano 2, o grupo tratado é comparado apenas com os nunca tratados e aqueles que ainda não foram tratados.
   - No Ano 4, o mesmo ocorre, mas agora o grupo tratado no Ano 2 não é mais usado como controle.

2. **Comparação em momentos relativos**:
   - Os resultados são apresentados como variações ao longo do tempo (antes, durante e depois do tratamento).



### Visualização

Um gráfico típico do estimador pode ser um **Event Study**, onde:

- O eixo horizontal representa o **tempo relativo ao tratamento**.
- O eixo vertical mostra os **efeitos estimados** ($ \delta_k $).

- **Antes do tratamento ($ k < 0 $)**: Deve haver efeitos próximos de zero, assumindo tendências paralelas.
- **Após o tratamento ($ k \geq 0 $)**: Observamos o impacto do tratamento ao longo do tempo.



### Limitações

1. **Especificação das interações**: A abordagem depende fortemente de uma boa especificação das interações.
2. **Controle de confundidores**: É essencial realizar um controle adequado de variáveis confundidoras.
3. **Complexidade de implementação**: Pode ser mais complexo de implementar em softwares padrão.

```python
# Load packages
import pandas as pd
import numpy as np
from datetime import date
from tqdm import tqdm
import plotly.graph_objects as go
import telegram_send
import paneleventstudy as es
from linearmodels import PanelOLS
import statsmodels.formula.api as smf
```

```python
df = pd.read_stata("https://github.com/Daniel-Uhr/data/raw/main/bacon_example.dta")
```

```python
# Outcome (Suicide Mortality)
df['Y'] = df['asmrs']
# Treatment D (Staggered treatment adoption)
df['D'] = df['post']
# Covariáveis - pcinc asmrh cases
df['X1'] = df['pcinc']
df['X2'] = df['asmrh']
df['X3'] = df['cases']

# Identificação do painel
df['year'] = df['year'].astype(int)
df['id'] = df['stfips'].astype('category').cat.codes + 1


# Relative time (Tempo relativo até o tratamento - ocorre somente para os tratados)
df['reltime'] = df['year'] - df['_nfd']
# colocar valor zero caso tenha dados faltantes
df['reltime'] = df['reltime'].fillna(0)
# garantir que 'reltime' seja um inteiro
df['reltime'] = df['reltime'].astype(int)

# Generates a new column called 'ct' containing the numerics version of calendar time 'year'
df['ct'] = df['year'] - 1964

# Vamos criar a variável de grupo (cohoort) e G (Callaway-SantAnna)
df['cohort'] = df['_nfd'] - 1965
df['G']=df['_nfd']

# control group
df['treated'] = 0
df.loc[df['_nfd'] > 0, 'treated'] = 1
df['control_group'] = 1
df.loc[df['treated'] == 1, 'control_group'] = 0

# Criar uma coluna que identifica os indivíduos que nunca foram tratados (nevertreated)
df['nevertreated'] = 0
df.loc[df['treated'] == 0, 'nevertreated'] = 1

# Criar uma coluna que identifica os indivíduos que sempre foram tratados (alwaystreated) Deve ser 1 para todos os anos após o tratamento D
df['mean_treat'] = df.groupby('id')['D'].transform('mean')
# Criar os sempre tratados
df['alwaystreated'] = 0
df.loc[(df['mean_treat'] == 1), 'alwaystreated'] = 1
```

```python
# TWFE - naive
smf.wls('asmrs ~ D + C(stfips) + C(year)', data=df).fit(cov_type='HC1').summary2().tables[1].loc[['Intercept', 'D'], :]
```

```python
# TWFE naive - Dynamic Difference-in-Differences

# Especificar a fórmula com a categoria 0 de reltime como referência diretamente
formula = 'Y ~ C(reltime, Treatment(reference=0)) + C(id) + C(year)'

# Ajustar o modelo
twfe = smf.ols(formula, data=df).fit(cov_type='HC1')

# Extraindo os resultados
beta = pd.DataFrame(twfe.params, columns=['coef'])  # Coeficientes estimados
ci = twfe.conf_int()  # Intervalos de confiança

# Renomear as colunas do intervalo de confiança para consistência
ci.columns = ['lower', 'upper']

# Combinar os resultados em um único DataFrame
est = beta.join(ci)

# Manter apenas os coeficientes relativos ao efeito de reltime
est = est.reset_index(drop=False)
est = est[est['index'].str.contains('C(reltime, Treatment(reference=0))[T.', regex=False)]  

# Clean the coefficient names for readability
est['index'] = est['index'].str.replace('C(reltime, Treatment(reference=0))[T.', '', regex=False)  # Remove the prefix
est['index'] = est['index'].str.replace(']', '', regex=False)         # Remove closing brackets

# Set the index to the cleaned relative time
est = est.set_index('index')

est = est.reset_index()  # Transforma o índice em uma coluna
est['index'] = pd.to_numeric(est['index'])  # Converte para numérico

est = est.sort_values('index')  # Agora ordenará corretamente
# Certifique-se de que o índice é numérico para ordenação
est.index = pd.to_numeric(est.index)
```

```python
import matplotlib.pyplot as plt

# Configurar o gráfico
plt.figure(figsize=(10, 6))

# Plotar os coeficientes com barras de erro
plt.errorbar(
    est['index'],  # Valores do eixo x
    est['coef'],   # Valores do eixo y (coeficientes)
    yerr=[est['coef'] - est['lower'], est['upper'] - est['coef']],  # Barras de erro
    fmt='o', color='blue', ecolor='gray', elinewidth=2, capsize=4, label='Estimated Coefficients'
)

# Adicionar uma linha horizontal no valor 0 para referência
plt.axhline(0, color='black', linestyle='--', linewidth=1)

# Adicionar título e rótulos
plt.title('Relative Time Coefficients with 95% Confidence Interval', fontsize=14)
plt.xlabel('Relative Time (index)', fontsize=12)
plt.ylabel('Coefficient Estimate (coef)', fontsize=12)

# Adicionar grade e legenda
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Exibir o gráfico
plt.show()
```

```python
dynamic_effects = []

for reltime in iw_estimates['reltime'].unique():
    # Filtrar coeficientes para o tempo relativo atual
    reltime_coefs = iw_estimates[iw_estimates['reltime'] == reltime]
    
    # Calcular efeito ponderado
    effect = sum(
        reltime_coefs['coef'] * reltime_coefs['cohort'].map(cohort_weights)
    )
    dynamic_effects.append({'reltime': reltime, 'Effect': effect})
```

```python
for row in dynamic_effects:
    reltime = row['reltime']
    reltime_coefs = iw_estimates[iw_estimates['reltime'] == reltime]
    
    # Variância ponderada
    variances = (reltime_coefs['coef'] ** 2) * (reltime_coefs['cohort'].map(cohort_weights) ** 2)
    weighted_var = variances.sum()
    std = np.sqrt(weighted_var)
    
    # Limites de confiança
    row['Lower Bound'] = row['Effect'] - 1.96 * std
    row['Upper Bound'] = row['Effect'] + 1.96 * std
```
